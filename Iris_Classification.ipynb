In order to give column names to our dataset, used the function iris_df.columns.

2. Passed the col_name list to iris_df.columns

```python
iris_df.columns = col_name
```

```python
iris_df.head(15)
```

In the above dataset, There are two kinds of attributes for modeling:

Continuous Variables : (These are numbers which can range from negative infinity to positive infinity. You can sort on such variables and filter by ranges) 
In the above dataset these are sepal_length, sepal_width, petal_length and petal_width.

Categorical Variables : (These variables can have a limited set of values, each of which indicate a sub-type) 
In the above data set, 'class' is a categorical variable because it can be either Iris-setosa, Iris Versicolour or Iris Virginica
